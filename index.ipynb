```python
import pandas as pd

df = pd.read_excel('./Car_Sales_Data.xlsx')

df_new = df.copy()

df_new.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# initialize LabelEncoder
le = LabelEncoder()

col_list = ['Gender', 'Model', 'Dealer_Name', 'Company', 'Engine', 'Transmission', 'Color', 'Body Style', 'Dealer_Region']

for col in col_list:
    df_new[col] = le.fit_transform(df_new[col].astype(str))

df_new.head()
```

```python
print(df_new['Date'].dtype)
# convert 'Date' column to datetime
df_new['Date'] = pd.to_datetime(df_new['Date'], errors='coerce')
df_new = df_new.dropna(subset=['Date'])
# extracting month, day, and year from 'Date'
df_new['month'] = df_new['Date'].dt.month
df_new['day'] = df_new['Date'].dt.day
df_new['year'] = df_new['Date'].dt.year
df_new.head()
```

```python
X = df_new[['Gender', 'Model', 'Annual Income', 'Dealer_Name', 'Company', 'Engine', 'Transmission', 'Color', 'Body Style', 'Dealer_Region', 'month', 'day', 'year']]
y = df_new['Price ($)']

print("Features shape:", X.shape)
print("Target shape:", y.shape)
```

```python
from sklearn.model_selection import train_test_split

# spliting the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=2203)

# displaying the shapes
print("Training features shape:", X_train.shape)
print("Testing features shape:", X_test.shape)
print("Training target shape:", y_train.shape)
print("Testing target shape:", y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# initialize the linear regression model
linear_model = LinearRegression()

# train
linear_model.fit(X_train, y_train)

# make predictions
y_pred_linear = linear_model.predict(X_test)

# print metrics
linear_r2 = r2_score(y_test, y_pred_linear)
linear_mse = mean_squared_error(y_test, y_pred_linear)
linear_mae = mean_absolute_error(y_test, y_pred_linear)

print("Linear Regression R-squared:", round(linear_r2, 4))
print("Linear Regression Mean Squared Error:", round(linear_mse, 4))
print("Linear Regression Mean Absolute Error:", round(linear_mae, 4))
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# initialize the decision tree model
tree_model = DecisionTreeRegressor(random_state=2203)

# train
tree_model.fit(X_train, y_train)

# make predictions
y_pred_tree = tree_model.predict(X_test)

# print metrics
tree_r2 = r2_score(y_test, y_pred_tree)
tree_mse = mean_squared_error(y_test, y_pred_tree)
tree_mae = mean_absolute_error(y_test, y_pred_tree)

print("Decision Tree R-squared:", round(tree_r2, 4))
print("Decision Tree Mean Squared Error:", round(tree_mse, 4))
print("Decision Tree Mean Absolute Error:", round(tree_mae, 4))
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# initialize the KNN model
knn_model = KNeighborsRegressor(n_neighbors=5)

# train
knn_model.fit(X_train, y_train)

# make predictions
y_pred_knn = knn_model.predict(X_test)

# print metrics
knn_r2 = r2_score(y_test, y_pred_knn)
knn_mse = mean_squared_error(y_test, y_pred_knn)
knn_mae = mean_absolute_error(y_test, y_pred_knn)

print("KNN Regressor R-squared:", round(knn_r2, 4))
print("KNN Regressor Mean Squared Error:", round(knn_mse, 4))
print("KNN Regressor Mean Absolute Error:", round(knn_mae, 4))
```

```python
from sklearn.ensemble import RandomForestRegressor

# initialize the Random Forest model
rf_model = RandomForestRegressor(n_estimators=100, random_state=2203)

# train
rf_model.fit(X_train, y_train)

# make predictions
y_pred_rf = rf_model.predict(X_test)

# print metrics
rf_r2 = r2_score(y_test, y_pred_rf)
rf_mse = mean_squared_error(y_test, y_pred_rf)
rf_mae = mean_absolute_error(y_test, y_pred_rf)

print("Random Forest R-squared:", round(rf_r2, 4))
print("Random Forest Mean Squared Error:", round(rf_mse, 4))
print("Random Forest Mean Absolute Error:", round(rf_mae, 4))
```
